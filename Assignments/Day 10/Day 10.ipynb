## Hypothesis Statements based on the following code from Spyder.

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 20 22:20:51 2020

@author: Akash Khyadi
"""

import pandas as pd

dataset=pd.read_csv("general_data.csv")

describe=dataset._get_numeric_data().describe()

count=dataset._get_numeric_data().count()

print(count)
print(describe)

#Displaying description of specific coulumns 
df=dataset[['Age','DistanceFromHome','MonthlyIncome','NumCompaniesWorked','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion']].describe()

#Skewness of the selected columns
sk=dataset[['Age','DistanceFromHome','MonthlyIncome','NumCompaniesWorked','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion']].skew()

#Kurtosis of the selected columns
kr=dataset[['Age','DistanceFromHome','MonthlyIncome','NumCompaniesWorked','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion']].kurt()

print(df)
print(sk)
print(kr)

#Check for Null Values
print(dataset.isnull())

#Drop rows with Null values 
nonull=dataset.dropna()
print("After dropping rows with null values",nonull)

#Check for duplicated values
print(nonull.duplicated())

#Drop duplicated rows from the dataset
nodup=nonull.drop_duplicates()
print("After dropping rows with duplicate values"+"\n",nodup)

#Adding cleaned data into a new dataset
df=nodup

#Dataset of nynumerical data values
df1=df[['Age','DistanceFromHome','Education','MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike','TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany','YearsSinceLastPromotion', 'YearsWithCurrManager']]

cleandes=df1.describe()

print("Describe= \n",cleandes,"\n")

print("Median= \n",df1.median(),"\n")

print("Mode= \n",df1.mode(),"\n")

print("Skew= \n",df1.skew(),"\n")

print("Kurtosis=\n",df.kurt(),"\n")

```
